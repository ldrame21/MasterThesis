# Bounding Box

```python
import requests
```

```python
# Coordinates defining the bounding box: [min x, min y, max x, max y]
# Replace these values with your actual coordinates in the Amersfoort / RD New (EPSG:7415) system
bbox_coordinates = [124451.42627059581, 487030.141436812, 125395.26047474254, 487351.62786675687]  # Example coordinates

api_endpoint = "https://api.3dbag.nl/collections/pand/items"

params = {
    'bbox': ','.join(map(str, bbox_coordinates)),
    'limit': 20
}

response = requests.get(api_endpoint, params=params)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}, {response.text}")
```

```python
# Assuming `data` contains the response data from the API
output_filename = 'data/out/buildings_data.json'

with open(output_filename, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

print(f"Data successfully saved to {output_filename}")
```

# Extract attribute from CityJSON file

## For 1 file

```python
# Load the JSON data from the file
filename = 'data/cityjson/8-440-720.city.json'

with open(filename, 'r', encoding='utf-8') as file:
    data = json.load(file)

# List of IDs you want to retrieve data for
ids = ["NL.IMBAG.Pand.0363100012061164", "NL.IMBAG.Pand.0363100012253622", "NL.IMBAG.Pand.0363100012252752", "NL.IMBAG.Pand.0363100012252737"]  # Add your 100 IDs here

# Dictionary to hold the retrieved data
retrieved_data = {}

# Iterate through the list of IDs and retrieve the values
for i, building_id in enumerate(ids):
    # Check if the ID exists in the CityObjects section of the JSON data
    print(f"Building {building_id} being processed...")
    if building_id in data['CityObjects']:
        print(f"Building {building_id} in data...")
        # Access the attributes section for this building ID
        attributes = data['CityObjects'][building_id]['attributes']
        
        # Retrieve 'b3_pw_datum' and 'b3_volume_lod12' values
        b3_pw_datum = attributes.get('b3_pw_datum')
        b3_volume_lod12 = attributes.get('b3_volume_lod12')
        
        # Store the retrieved values in the dictionary
        retrieved_data[building_id] = {
            'b3_pw_datum': b3_pw_datum,
            'b3_volume_lod12': b3_volume_lod12
        }

# Now 'retrieved_data' contains the data for all the IDs you were interested in
# For example, to print the data for an ID:
print(retrieved_data)
```

## For a folder

```python
# Assuming 'ids' is a set of your 30,000 IDs
ids = {"NL.IMBAG.Pand.0363100012061164", "NL.IMBAG.Pand.0363100012253622", "NL.IMBAG.Pand.0363100012252752", 
       "NL.IMBAG.Pand.0363100012252737", "NL.IMBAG.Pand.0363100012061172", "NL.IMBAG.Pand.0363100012061848"} # Add your 100 IDs here

# Directory containing your city.json files
directory = 'data/cityjson/'

# Dictionary to store the retrieved data
retrieved_data = {}

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.city.json'):
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
            print(f"Processing {filename}...")
            data = json.load(file)

            # Check each ID in the set against this file's data
            for building_id in list(ids):  # Use list to avoid RuntimeError for set size change during iteration
                if building_id in data['CityObjects']:
                    print(f"Building {building_id} being processed...")
                    attributes = data['CityObjects'][building_id]['attributes']
                    b3_pw_datum = attributes.get('b3_pw_datum')
                    b3_volume_lod12 = attributes.get('b3_volume_lod12')

                    # Store the retrieved data
                    retrieved_data[building_id] = {
                        'b3_pw_datum': b3_pw_datum,
                        'b3_volume_lod12': b3_volume_lod12
                    }

                    # Remove the found ID from the set to avoid unnecessary future searches
                    ids.remove(building_id)

            # Optional: Early exit if all IDs are found
            if not ids:
                break

# At this point, 'retrieved_data' contains the data for all found IDs
```
