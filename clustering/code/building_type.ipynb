```python
df_ams = df[['building_id','year_of_construction', 'postal_code', 'house_number', 'house_letter', 'house_number_addition',
             'neighborhood_name', 'house_type']]
df_ams
```

```python
df_ams['house_number'] = df_ams['house_number'].fillna(0).astype(int)
df_ams['postal_code'] = df_ams['postal_code'].fillna('0')
df_ams['house_letter'] = df_ams['house_letter'].fillna('0')
df_ams['house_number_addition'] = df_ams['house_number_addition'].fillna(0)

df_ams['UID'] = df_ams['postal_code'].astype(str) + '-' + df_ams['house_number'].astype(str) #+ '-' + df_ams['house_letter'].astype(str) + '-' + df_ams['house_number_addition'].astype(str)
df_ams
```

```python
#Loading EPC data
DATA_PATH = '../data/out/dataset_v2024520164517'
df_epc = pd.read_csv(DATA_PATH)
df_epc
```

```python
df_epc['Building_house_number'] = df_epc['Building_house_number'].fillna(0).astype(int)
df_epc['Building_postcode'] = df_epc['Building_postcode'].fillna('0')
df_epc['Building_house_letter'] = df_epc['Building_house_letter'].fillna('0')
df_epc['Building_house_number_addition'] = df_epc['Building_house_number_addition'].fillna(0)

df_epc['UID'] = df_epc['Building_postcode'].astype(str) + '-' + df_epc['Building_house_number'].astype(str) #+ '-' + df_epc['Building_house_letter'].astype(str) + '-' + df_epc['Building_house_number_addition'].astype(str)
df_epc
```

```python
# Join BAG and EPC data
merged_df = df_ams.merge(df_epc, on='UID', how='left')

# Retrieve rows from df1 that were not joined (i.e., where df2 columns are NaN)
unmatched_rows = merged_df[merged_df['postal_code'].isna()]

unmatched_rows
```

```python
out = merged_df[['building_id', 'house_type', 'Building_type', 'Building_energy_class', 'postcode', 'postcode_num']]
out.dropna(subset='building_id', inplace=True)
out
```

```python
# Applying the conditions
# Replace NaN in 'house_type' with the value in 'Building_type' when 'Building_type' is not NaN
out.loc[out['house_type'].isna() & out['Building_type'].notna(), 'house_type'] = out['Building_type']

# Replace 'Appartment' in 'house_type' with the value in 'Building_type' when 'Building_type' is not NaN
out.loc[(out['house_type'] == 'Appartement') & out['Building_type'].notna(), 'house_type'] = out['Building_type']

out['house_type'].isna().sum(), out[out['house_type'] == 'Appartement'].shape[0]
```

```python
# Function to get the most common Building_energy_class per building_id
def most_common(series):
    return series.mode().iloc[0] if not series.mode().empty else None

# Find the most common Building_energy_class for each building_id
common_energy_class = out.groupby('building_id')['Building_energy_class'].agg(most_common).reset_index()
common_energy_class
```

```python
# Merge to get the rows with the most common Building_energy_class
result_df = pd.merge(out, common_energy_class, on=['building_id', 'Building_energy_class'])
result_df
```

```python
# Function to apply to each group
def filter_rows_maisonnette(group):
    if 'Maisonnette' in group['house_type'].values:
        non_nan_appartement = group['house_type'].dropna().unique()
        if len(non_nan_appartement) == 1 and non_nan_appartement[0] == 'Maisonnette':
            # If all types are 'apartment', return the first one
            return group.iloc[:1]
        else:
            # Return all but 'apartment' type rows if not all are 'apartment'
            return group[group['house_type'] != 'Maisonnette']
    return group

# Function to apply to each group
def filter_rows_flat(group):
    if 'Flatwoning (overig)' in group['house_type'].values:
        non_nan_appartement = group['house_type'].dropna().unique()
        if len(non_nan_appartement) == 1 and non_nan_appartement[0] == 'Flatwoning (overig)':
            # If all types are 'apartment', return the first one
            return group.iloc[:1]
        else:
            # Return all but 'apartment' type rows if not all are 'apartment'
            return group[group['house_type'] != 'Flatwoning (overig)']
    # Return all rows if 'apartment' does not exist
    return group

# Function to apply to each group
def filter_rows_appartment(group):
    if 'Appartement' in group['house_type'].values:
        non_nan_appartement = group['house_type'].dropna().unique()
        if len(non_nan_appartement) == 1 and non_nan_appartement[0] == 'Appartement':
            # If all types are 'apartment', return the first one
            return group.iloc[:1]
        else:
            # Return all but 'apartment' type rows if not all are 'apartment'
            return group[group['house_type'] != 'Appartement']
    # Return all rows if 'apartment' does not exist
    return group

# Apply the function to each group and concatenate the results
filtered_df = result_df.groupby('building_id').apply(filter_rows_maisonnette).reset_index(drop=True)
# Apply the function to each group and concatenate the results
filtered_df = filtered_df.groupby('building_id').apply(filter_rows_flat).reset_index(drop=True)
# Apply the function to each group and concatenate the results
filtered_df = filtered_df.groupby('building_id').apply(filter_rows_appartment).reset_index(drop=True)
```
