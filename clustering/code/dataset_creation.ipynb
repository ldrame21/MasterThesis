# Dataset Preparation

```python
df_residence = df_ams[['Building_bag_building_id','Building_type','Building_subtype','Building_postcode_num','Building_energy_class', 'Building_heat_demand', 'Building_share_renewable_energy', 'Building_primary_fossil_energy', 'Building_energy_requirement',
                          'Building_usable_floor_area_thermal_zone']].dropna(subset=['Building_bag_building_id','Building_energy_class', 'Building_heat_demand', 'Building_share_renewable_energy', 'Building_primary_fossil_energy', 'Building_energy_requirement',
                          'Building_usable_floor_area_thermal_zone'])
df_residence
```

```python
df_residence['id'] = "NL.IMBAG.Pand.0" + df_residence['Building_bag_building_id'].astype('int64').astype(str)
df_residence
```

```python
# Assuming 'ids' is a set of your 30,000 IDs
ids = set(df_residence['id'].unique().tolist())

# Directory containing your city.json files
directory = 'data/cityjson/'

# Dictionary to store the retrieved data
retrieved_data = {}

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.city.json'):
        with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
            print(f"Processing {filename}...")
            data = json.load(file)

            # Check each ID in the set against this file's data
            for building_id in list(ids):  # Use list to avoid RuntimeError for set size change during iteration
                if building_id in data['CityObjects']:
                    #print(f"Building {building_id} being processed...")
                    attributes = data['CityObjects'][building_id]['attributes']
                    oorspronkelijkbouwjaar = attributes.get('oorspronkelijkbouwjaar')
                    b3_volume_lod22 = attributes.get('b3_volume_lod22')
                    b3_bouwlagen = attributes.get('b3_bouwlagen')
                    b3_dak_type = attributes.get('b3_dak_type')
                    b3_kas_warenhuis = attributes.get('b3_kas_warenhuis')
                    geographical_extent = data['CityObjects'][building_id].get('geographicalExtent', [])

                    # Store the retrieved data
                    retrieved_data[building_id] = {
                        'id': building_id,
                        'oorspronkelijkbouwjaar': oorspronkelijkbouwjaar,
                        'b3_volume_lod22': b3_volume_lod22,
                        'b3_bouwlagen': b3_bouwlagen,
                        'b3_dak_type': b3_dak_type,
                        'b3_kas_warenhuis': b3_kas_warenhuis,
                        'geographicalExtent': geographical_extent
                        
                    }

                    # Remove the found ID from the set to avoid unnecessary future searches
                    ids.remove(building_id)

            # Optional: Early exit if all IDs are found
            if not ids:
                break

# At this point, 'retrieved_data' contains the data for all found IDs
```
