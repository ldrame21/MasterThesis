```python
import pandas as pd
import json
import os
import numpy as np
import datetime
```

# Data Loading

```python
try:
    df = pd.read_csv('../data/v20240401_v2_csv.csv', sep=';')
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df.shape
```

```python
#Rename column in english
df.columns = [
    'Building_admission_date',  # Pand_opnamedatum
    'Building_admission_type',  # Pand_opnametype
    'Building_status',  # Pand_status
    'Building_calculation_type',  # Pand_berekeningstype
    'Building_energy_index',  # Pand_energieindex
    'Building_energy_class',  # Pand_energieklasse
    'Building_energy_label_is_private',  # Pand_energielabel_is_prive
    'Building_based_on_reference_building',  # Pand_is_op_basis_van_referentie_gebouw
    'Building_class',  # Pand_gebouwklasse
    'Measurement_valid_until',  # Meting_geldig_tot
    'Building_registration_date',  # Pand_registratiedatum
    'Building_postcode',  # Pand_postcode
    'Building_house_number',  # Pand_huisnummer
    'Building_house_letter',  # Pand_huisletter
    'Building_house_number_addition',  # Pand_huisnummertoevoeging
    'Building_detail_designation',  # Pand_detailaanduiding
    'Building_bag_residence_object_id',  # Pand_bagverblijfsobjectid
    'Building_bag_mooring_place_id',  # Pand_bagligplaatsid
    'Building_bag_standing_place_id',  # Pand_bagstandplaatsid
    'Building_bag_building_id',  # Pand_bagpandid
    'Building_type',  # Pand_gebouwtype
    'Building_subtype',  # Pand_gebouwsubtype
    'Building_project_name',  # Pand_projectnaam
    'Building_project_object',  # Pand_projectobject
    'Building_SBI_code',  # Pand_SBIcode
    'Building_usable_floor_area_thermal_zone',  # Pand_gebruiksoppervlakte_thermische_zone
    'Building_energy_requirement',  # Pand_energiebehoefte
    'Building_requirement_energy_requirement',  # Pand_eis_energiebehoefte
    'Building_primary_fossil_energy',  # Pand_primaire_fossiele_energie
    'Building_requirement_primary_fossil_energy',  # Pand_eis_primaire_fossiele_energie
    'Building_primary_fossil_energy_EMG_standard',  # Pand_primaire_fossiele_energie_EMG_forfaitair
    'Building_share_renewable_energy',  # Pand_aandeel_hernieuwbare_energie
    'Building_requirement_share_renewable_energy',  # Pand_eis_aandeel_hernieuwbare_energie
    'Building_share_renewable_energy_EMG_standard',  # Pand_aandeel_hernieuwbare_energie_EMG_forfaitair
    'Building_temperature_exceedance',  # Pand_temperatuuroverschrijding
    'Building_requirement_temperature_exceedance',  # Pand_eis_temperatuuroverschrijding
    'Building_heat_demand',  # Pand_warmtebehoefte LDD
    'Building_energy_index_with_EMG_standard'  # Pand_energieindex_met_EMG_forfaitair
]

#Add registration year
df['Building_registration_year'] = df['Building_registration_date'].apply(lambda date: int(str(date)[:4]))
#Add Postcode number 
df = df.dropna(subset=['Building_postcode'])
df['Building_postcode_num'] = df['Building_postcode'].apply(lambda postcode: int(str(postcode)[:4]))
df
```

```python
#Create ams dataset (11-- postcodes still need to be added)
df_ams = df[df['Building_postcode_num'].isin(list(range(1000, 1113, 1)))]
df_ams
```

```python
selected_columns = num_columns + ['Building_type'] + ['Building_subtype'] + ['Building_energy_class'] + ['Building_postcode'] + ['Building_house_number_addition'] + ['Building_house_letter']
```

```python
df_ams = df_ams[selected_columns]
```

```python
gb = df_ams.groupby('Building_bag_residence_object_id')[['Building_registration_year', 'Building_energy_requirement', 
                                                         'Building_primary_fossil_energy', 'Building_heat_demand', 
                                                         'Building_requirement_share_renewable_energy']].max()
#gb = gb.dropna()
gb
```

```python
gb2 = df_ams.groupby('Building_registration_year')[['Building_registration_year', 'Building_energy_requirement', 
                                                         'Building_primary_fossil_energy', 'Building_heat_demand', 'Building_requirement_share_renewable_energy']].count()
gb2
```

```python
gb3 = df_ams[df_ams.Building_registration_year >= 2021].groupby('Building_postcode_num')[['Building_registration_year', 'Building_energy_requirement', 
                                                         'Building_primary_fossil_energy', 'Building_heat_demand', 'Building_requirement_share_renewable_energy']].count()
gb3
```
