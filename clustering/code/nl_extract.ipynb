```python
# Translation dictionary
translation_dict = {
    'Appartement': 'Apartment',
    'Tussen of geschakelde woning': 'Terraced or semi-detached house',
    'Vrijstaande woning': 'Detached house',
    'Hoekwoning': 'Corner house',
    'Tweeonder1kap': 'Semi-detached house',
    np.nan: np.nan  # To handle NaN values
}

# Translation function
def translate_property_type(property_type):
    return translation_dict.get(property_type, property_type)

# Apply the function to the DataFrame
df['property_type'] = df['woningtype'].apply(translate_property_type)
```
